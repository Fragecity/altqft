```python
import numpy as np
from qiskit import QuantumCircuit
from qiskit.quantum_info import Operator
```

```python
qc = QuantumCircuit(4)
qc.h(1)
qc.cp(1.5, 1,0)
qc.cp(2,1,2)
qc.h(0)
qc.cp(3, 0,2)
qc.cp(1.7, 0,3)
qc.h(2)
qc.cp(2.5,2,3)
qc.h(3)


print(qc)
```

```python
qc_mat = np.array(Operator(qc).data )
qc_mat = qc_mat/qc_mat[0,0]

vec = qc_mat[:,5]
vec
```

```python
vec1 = vec[slice(0,-1, 2)]
print(vec1)
vec2 = vec[slice(1,-1, 2)]
print(vec2 / vec2[0])
```

```python
qc = QuantumCircuit(4)
qc.h(0)
qc.h(2)
qc.cp(1, 0,3)
qc.cp(1.5, 1,2)
qc.cp(1.8, 0,1)
qc.cp(2, 2,3)
qc.h(1)
qc.h(3)

print(qc)
```

```python
qc_mat = np.array(Operator(qc).data )
qc_mat = qc_mat/qc_mat[0,0]

vec = qc_mat[:,3]
vec1 = vec[slice(0,-1, 2)]
print(vec1)
vec2 = vec[slice(1,-1, 2)]
print(vec2 / vec2[0])
```
